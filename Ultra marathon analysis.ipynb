```python
# Importing Libraries

import pandas as pd
import seaborn as sns
```

```python
# Loading Dataset

df = pd.read_csv("TWO_CENTURIES_OF_UM_RACES.csv")
df.head(10)
```

```python
#Checking datatypes and shape of Dataset

df.shape
df.dtypes
```

```python
# Cleaning up data
# Only want USA Races, distances that are 50k or 50mi, year 2020
# Filtering Dataset below
```

```python
# Races with distance of 50km

df[df['Event distance/length'] == '50km']
```

```python
# Races with distance of 50mi

df[df['Event distance/length'] == '50mi']
```

```python
# Combining 50k/50mi with isin

df[df['Event distance/length'].isin(['50km','50mi'])]
```

```python
# Combining distance filter with races in the year 2020

df[(df['Event distance/length'].isin(['50km','50mi']))&(df['Year of event']==2020)]
```

```python
# Checking for Races in the USA

df[df['Event name']=='Everglades 50 Mile Ultra Run (USA)']['Event name'].str.split('(').str.get(1).str.split(')').str.get(0)
```

```python
# Filtering races in the USA

df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0)=='USA']
```

```python
#Combining all filters together

df[(df['Event distance/length'].isin(['50km','50mi']))&(df['Year of event']==2020)
   &(df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0)=='USA')]
```

```python
# Creating a seperate Dataframe

df2 = df[
    (df['Event distance/length'].isin(['50km', '50mi'])) &
    (df['Year of event'] == 2020) &
    (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')
]

df2.head()
```

```python
df2.shape
```

```python
# Removing (USA) from event names

df2['Event name'].str.split('(').str.get(0)
```

```python
# Applying Filter

df2['Event name']=df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head()
```

```python
# Adding a new column 'athlete_age'

df2['athlete_age']=2020-df2['Athlete year of birth']
```

```python
# Removing h from athlete performance

df2['Athlete performance']=df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head()
```

```python
# dropping columns 'Athlete club','Athlete country','Athlete year of birth','Athlete age category'


df2 = df2.drop(['Athlete club','Athlete country','Athlete year of birth','Athlete age category'],axis = 1)
df2.head()
```

```python
# Clean up nulls

df2.isna().sum()
```

```python
df2[df2['athlete_age'].isna()==1]
```

```python
df2 = df2.dropna()
df2.shape
```

```python
# Check for duplicates

df2[df2.duplicated()==True]
```

```python
# Resetting index

df2.reset_index(drop = True)
```

```python
# Fixing datatypes

df2.dtypes
```

```python
# Applying filters to fix datatypes

df2['athlete_age'] = df2['athlete_age'].astype(int)
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
df2.dtypes
```

```python
# Renaming column names

df2 = df2.rename(columns = {
    'Year of event': 'year',
    'Event dates': 'event_dates',
    'Event name':'race_name',
    'Event distance/length':'race_length',
    'Event number of finishers':'race_number_of_finishers',
    'Athlete performance':'athlete_performance',
    'Athlete gender':'athlete_gender',
    'Athlete average speed': 'athlete_average_speed',
    'Athlete ID':'athlete_id',
    'athlete_age':'athlete_age'
})
```

```python
df2.head()
```

```python
# Reordering for better understanding

df3 = df2[['event_dates', 'race_name', 'race_length', 'race_number_of_finishers',
           'athlete_id', 'athlete_gender', 'athlete_age', 'athlete_performance', 'athlete_average_speed',
             'year']]
```

```python
df3.head()
```

```python
# Charts and Graphs
```

```python
# 50km vs 50mi races

sns.histplot(df3['race_length'])
```

```python
# Males vs Females

sns.histplot(df3, x= 'race_length', hue = 'athlete_gender')
```

```python
sns.histplot(df3, x='race_length', hue = 'athlete_gender')
```

```python
# Average athlete speed

sns.displot(df3[df3['race_length']=='50mi']['athlete_average_speed'])
```

```python
# Distribution of average speeds for a given race length, split by gender.

sns.violinplot(data = df3, x='race_length', y='athlete_average_speed', hue = 'athlete_gender', split = True, inner = 'quart', linewidth=1)
```

```python
# Average speed decreases with age for both genders

sns.lmplot(data=df3, x='athlete_age', y='athlete_average_speed', hue='athlete_gender')
```

```python
# Differnce in speed for the 50k, 50mi male to female

df3.groupby(['race_length','athlete_gender'])['athlete_average_speed'].mean()
```

```python
# What age groups are the best in the 50m Race Top 10

df3.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = False).query('count>19').head(10)
```
